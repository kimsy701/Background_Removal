# A. 배경제거 하기 전 이미지 Augmentation


```python
in_input_path = "smoretalk/ft_model_in_aug/input_aug/"
in_output_path = "smoretalk/ft_model_in_aug/output_aug/"
```

```python
# flip anti-clockwise
ori_path = os.listdir(in_input_path)

ori_files = [ori_img for ori_img in ori_path if ori_img.lower().endswith((".jpg", ".png", ".jpeg", ".gif"))]


for i,ori_img in enumerate(ori_files):
  ori_path = os.path.join(in_input_path, ori_img)
  # Check if the image has an alpha channel (transparency)

  ori_image = Image.open(ori_path)


  flip_img = ori_image.transpose(Image.TRANSPOSE)
  flip_img.save(os.path.join(in_output_path, "flip_anti_cl",ori_img))

  flip_img = ori_image.transpose(Image.TRANSVERSE)
  flip_img.save(os.path.join(in_output_path, "flip_cl",ori_img))

  flip_img = ori_image.transpose(Image.FLIP_LEFT_RIGHT)
  flip_img.save(os.path.join(in_output_path, "flip_hori",ori_img))

  flip_img = ori_image.transpose(Image.FLIP_TOP_BOTTOM)
  flip_img.save(os.path.join(in_output_path, "flip_verti",ori_img))

  flip_img = ori_image.transpose(Image.ROTATE_90)
  flip_img.save(os.path.join(in_output_path, "rot_90",ori_img))

  flip_img = ori_image.transpose(Image.ROTATE_180)
  flip_img.save(os.path.join(in_output_path, "rot_180",ori_img))

  flip_img = ori_image.transpose(Image.ROTATE_270)
  flip_img.save(os.path.join(in_output_path, "rot_270",ori_img))
```
