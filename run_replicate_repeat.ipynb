```python
#10분 주기
import replicate
import time


for i in range(3):

  a= time.time()
  output = replicate.run(
    "smoretalk/ispyrenet_vit_a40:240ac47394516f4447f07dac5f3edfa862c9c09d6185e770cef155938ae96318",
    input={"image": "https://i.ibb.co/kXsn19j/an-image-of-a-Rubik-s-Cube-1.png"}
  )
  b=time.time()
  print(i+1,"번째 시간:",b-a)
  time.sleep(60*10)

#첫번째 전에 cold start 안걸렸을수도 있음
```
